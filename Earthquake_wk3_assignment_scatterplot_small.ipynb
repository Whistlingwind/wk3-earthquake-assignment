```python
#Purpose of this file: Zoom in on interesting countries for further analysis of the magnitude and depth of earthquakes
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

#Read CSV data
df = pd.read_csv(r'https://raw.githubusercontent.com/Whistlingwind/wk3-earthquake-assignment/main/modified_all_month.csv')

#Create a country level index for future handling
countryIndex_ = df['country']

#Override df dataframe indexes to allow for aggregation
df.index = countryIndex_

#Create a new dataframe, and requesting specific countries via their index
df2 = df.loc[['Alaska','California','Hawaii','Washington','Nevada','Puerto Rico']]

#Create meta data, such as number of subplots and their sizes, as well as if they are using the same x & y values (They are), (I was going to make more subplots but then couldn't be bothered due to time)
fig, axes = plt.subplots(1, 1, figsize=(5, 5), sharey=True)
fig.suptitle('Top Countries/States with Earthquakes')

#Create the scatterplot with magnitude, depth and by country as coloured
sns.scatterplot(x=df2["mag"], y=df2["depth"],hue=df2["country"])

#Move legend to the top right, because it was populating inside the graph blocking the data view
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
