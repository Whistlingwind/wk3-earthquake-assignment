```python
#Purpose of the file: Show quantitative measure across country / state level abstraction, to identify key locations for further analysis

import plotly.express as px
import datetime as dt
import pandas as pd
import dateutil
 
# reading the csv file
df = pd.read_csv(r'https://raw.githubusercontent.com/Whistlingwind/wk3-earthquake-assignment/main/modified_all_month.csv')

#Create index to allow for aggregation across countries
countryIndex_ = df['country']

#Apply index to the dataframe and override any prior index
df.index = countryIndex_

#Using the indexing created, select indexed elements that I want to load into the histogram
df2 = df.loc[['Alaska','California','Hawaii','Washington','Nevada','Puerto Rico', 'Oklahoma','Montana','Texas','New Mexico','Indonesia', 'Japan', 'Utah']]

#Create histogram, pulling in country data from the newly created subset dataframe df2
fig = px.histogram(df2, y=df2['country'], color=df2['country'],
                   labels=dict(country="Top Affected Countries"), 
                   width=1000,height=500 ).update_yaxes(categoryorder='total ascending')

# showing the plot
fig.show()
```
