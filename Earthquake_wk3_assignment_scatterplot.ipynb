```python
#Purpose of file: To analyse full set of data, identifying magnitude and depth impact, and help isolate interesting trends or stats
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

#Read CSV data
countries = pd.read_csv(r'https://raw.githubusercontent.com/Whistlingwind/wk3-earthquake-assignment/main/modified_all_month.csv')

#create a sample, incase the load crashes or takes to long to process
sample = countries.sample(9000)

#Create scatter plot based on magnitude and depth, with colur referencing country. Using the sampled dataset
sns.scatterplot(sample["mag"], sample["depth"], hue=sample["country"])
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
